```python
import pandas as pd
path = "C:\\Users\\Othamne\\OneDrive - University of Ottawa\\Winter 2023\\CSI 4107\\Assignment 1\\Fundamentals-of-Data-Science\\Physical-Design-Deliverable-2\\assets"
Game = pd.read_csv(path+"\\game\\game_dimension.csv", index_col=0)
Sales = pd.read_csv(path+"\\game\\vgsales.csv", index_col=0)
Pub =pd.read_csv(path+"\\publisher\\Publisher-T.csv")
Con = pd.read_csv(path+'\\console\\console.csv', index_col=0)
```

## Merging the dimensions

```python
tmp = pd.merge(Sales, Game, left_on=['Name','Year'], right_on=['Name','Release_date'], how = 'inner')[Sales.columns]
```

```python
tmp2 = pd.merge(tmp, Con, left_on=['Platform'], right_on=['ConsoleID'], how = 'inner')[Sales.columns]
```

```python
merged = pd.merge(Pub,tmp2,  left_on='P_name', right_on='Publisher', how = 'inner')[Sales.columns]
```

```python
merged.to_csv(path+"//Sales_table.csv")
```
